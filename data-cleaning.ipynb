```python
import pandas as pd
import numpy as np
from pandas_datareader import data
```

```python
listing_df = pd.read_json('etsy-painting-listings.json')
pd.set_option('display.max_columns', listing_df.shape[1])
#drop irrelevant columns
listing_df = listing_df.drop(['state','user_id','category_id','title','description','creation_tsz','ending_tsz',
                              'original_creation_tsz','last_modified_tsz','sku','shop_section_id','featured_rank',
                              'state_tsz','url','views','num_favorers','shipping_template_id','processing_min',
                              'processing_max','recipient','occasion','non_taxable','file_data','should_auto_renew',
                              'language','taxonomy_id','is_private'], axis=1)
listing_df
```

```python
#check if there is any duplicate of listing id
print(listing_df.duplicated(subset=['listing_id']).sum(),'duplicates')
#extract duplicates
listing_df.loc[listing_df.duplicated(subset=['listing_id'])]
tmp_filt = (listing_df['listing_id']==670659796)
listing_df.loc[tmp_filt]
#keep the first instance
listing_df.drop_duplicates(subset=['listing_id'],inplace=True)

#find null values
print(listing_df.isnull().sum())
```

```python
#drop style because too many null
listing_df.drop(['style'],axis=1,inplace=True)

#set listing_id as index
listing_df.set_index('listing_id',inplace=True)
listing_df.head(10)
```

```python
#price column
#check all unique values 
listing_df['currency_code'].value_counts()
#gather info about these currencies and all convert to USD
xr_dict = dict()
xr_dict['USD']=1.0
xr_dict['GBP']=data.DataReader('DEXUSUK', 'fred').iloc[-1].values[0]
xr_dict['EUR']=data.DataReader('DEXUSEU', 'fred').iloc[-1].values[0]
xr_dict['CAD']=1/data.DataReader('DEXCAUS', 'fred').iloc[-1].values[0]
xr_dict['AUD']=data.DataReader('DEXUSAL', 'fred').iloc[-1].values[0]
xr_dict['TRY']=0.12
xr_dict['ILS']=0.31
xr_dict['NZD']=data.DataReader('DEXUSNZ', 'fred').iloc[-1].values[0]
xr_dict['DKK']=1/data.DataReader('DEXDNUS', 'fred').iloc[-1].values[0]
xr_dict['SEK']=1/data.DataReader('DEXSDUS', 'fred').iloc[-1].values[0]
xr_dict['HKD']=1/data.DataReader('DEXHKUS', 'fred').iloc[-1].values[0]
xr_dict['SGD']=1/data.DataReader('DEXSIUS', 'fred').iloc[-1].values[0]
xr_dict['MXN']=1/data.DataReader('DEXMXUS', 'fred').iloc[-1].values[0]
xr_dict['NOK']=1/data.DataReader('DEXNOUS', 'fred').iloc[-1].values[0]
def convert_curr(row):
    curr_name = row['currency_code']
    rate = xr_dict[curr_name]
    return rate
listing_df['xr']=listing_df.apply(convert_curr, axis=1)    
listing_df['price']=round(listing_df['price']*listing_df['xr'],2)
listing_df.head(10)
#drop unnecessary cols
listing_df = listing_df.drop(['currency_code','xr','quantity','tags'],axis=1)
listing_df.head()
```

```python
#explore the rest of the cols
listing_df['who_made'].value_counts()
listing_df['is_supply'].value_counts()
listing_df['when_made'].value_counts()
listing_df['is_customizable'].value_counts()
listing_df['is_digital'].value_counts()
listing_df['has_variations'].value_counts()
listing_df['taxonomy_path'].head(10)
listing_df['used_manufacturer'].value_counts()
listing_df['is_vintage'].value_counts()
np.isnan(listing_df['item_length'].iloc[0])
```

```python
#convert dimension into mm
dim_filt = (listing_df['item_dimensions_unit']=='in')
listing_df.loc[dim_filt, ['item_length','item_width','item_height']] = \
    listing_df.loc[dim_filt, ['item_length','item_width','item_height']]*25.4
```

```python
listing_df = listing_df.drop(['is_supply','item_weight','item_weight_unit', 'taxonomy_path','used_manufacturer',
                 'item_dimensions_unit'],axis=1)
listing_df.columns
```
